```python
id=df['ID'][0]
columm='SCHOOL TYPE'
df[df['ID']==id][columm]
```

```python
def append_entries_in_one(id, columm):
    d = {}
    d['items'] = []
    d['wanted'] = 0
    for item in df[df['ID']==id][columm]:
        d['items'].append(item)
        if item in ["Ungraded", "High School", "Middle School", "Elementary School", "Junior High School"]:
            d['wanted'] +=1
    d.update({'len': len(d['items'])})  
    d.update({'id': id})  
    
    return (d['wanted'], d['len'], d['items'], d['id'])
```

```python
append_entries_in_one(df['ID'][10], 'SCHOOL TYPE')
```

```python
df['ID'].loc[9]
```

```python
d = df.apply(lambda x: append_entries_in_one(x['ID'], 'SCHOOL TYPE'), axis=1)
```
